## Functions

```python
# write a function to that generates a normalized 1-d Gaussian, given mu, sigma
# plot the output for a few combinations of mu,sigma. Plot a vertical line at the mean.
```

```python
def gauss(x, mu=0, sig=1):
    '''Gaussian function'''
    if (sig<=0):
        print("sigma must be positive definite")
        return -1
    y=(x-mu)**2/(2*sig**2)
    y=np.exp(-y)/np.sqrt(2*np.pi*sig**2)
    return y        
```

```python
x=np.arange(-3,3, 0.01)
y=gauss(x, 0, -1)
y
```

```python
mu1=1
sig1=2
x=np.arange(-10,10, 0.01)
y=gauss(x, mu1, sig1)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, y,"k:")
plt.xlabel("x")
plt.ylabel("y")
plt.vlines(mu1, 0, 1/np.sqrt(2*np.pi*sig1**2), color="r", label=r"$\mu$")
plt.legend()
```

## Dictionaries

```python
D={}
D["mu"]=np.array([2,3])
D["sig"]=np.array([0.4, 1])
D["A"]=np.array([-1, 2.5])
```

```python
# write a function that takes a dictionary with arrays of mu, sig, amplitude A as inputs
# and returns the sum of the corresponding Gaussians
# add the required sanity checks
```

```python
def gauss_mix(x,D):
    mu=D["mu"]
    sig=D["sig"]
    A=D["A"]
    y=np.exp(-(x-mu[:, None])**2/(2*sig[:,None]**2))
    y*=A[:,None]/np.sqrt(2*np.pi*sig[:,None]**2)
    y=np.sum(y, axis=0)
    return y
    
    
```

```python
def gauss_mix1(x,D):
    mu=D["mu"]
    sig=D["sig"]
    A=D["A"]
    y=np.zeros_like(x)
    for i in range(len(mu)):
        y+=A[i]*gauss(x, mu[i], sig[i])
    return y
    
```

```python
y1=gauss_mix(x, D)
y2=gauss_mix1(x, D)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, y1, "r")
plt.plot(x, y2, "k")
plt.xlabel("x")
plt.ylabel("y")
```
