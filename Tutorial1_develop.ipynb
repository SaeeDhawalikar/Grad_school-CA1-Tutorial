# Contour plots

```python
data = np.loadtxt("data/2D_gaussian.dat", delimiter=',')

plt.imshow(data, origin="lower", cmap="Blues")
plt.contour(data, levels=2, colors='C0')
plt.show()
```

```python
data.shape
```

```python
N = data.shape[0]
x = np.linspace(-3, 3, N)
y = np.linspace(-3, 3, N)
X, Y = np.meshgrid(x, y)

plt.imshow(data, extent=[-3,3,-3,3], origin="lower", cmap="Blues")
plt.contour(X, Y, data, levels=2, colors='tab:blue')
plt.show()
```

```python
plt.contourf(X, Y, data, levels=5, cmap='Blues')
plt.colorbar()
plt.show()
```

```python
# for sigma contours,
# you need to compute the probabilities where the value crosses the particular sigma values

# Flatten and sort data in descending order
Z_sorted = np.sort(data.ravel())[::-1]
Z_cumsum = np.cumsum(Z_sorted)
Z_cumsum /= Z_cumsum[-1]  # normalization

# Function to get PDF threshold for given probability
def find_level(prob):
    return Z_sorted[np.searchsorted(Z_cumsum, prob)]

level_1sigma = find_level(0.6827)
level_2sigma = find_level(0.9545)


# plotting
plt.imshow(data, extent=[-3,3,-3,3], origin="lower", cmap="Blues")
cs = plt.contour(X, Y, data, levels=[level_2sigma, level_1sigma], colors="C0")
txt_objects = plt.clabel(cs, fmt="%0.3f", fontsize=10)
txt_objects[0].set(text=r"$2\sigma$")
txt_objects[1].set(text=r"$1\sigma$")
plt.show()
```

# Colour coded curves

```python
x=np.linspace(0, np.pi, 100)
A=np.arange(-3,4,1)
y=np.zeros([len(A), len(x)])
for i in range(len(A)):
    y[i]=np.sin(x*A[i])
A
```

```python
for i in range(len(y)):
    plt.plot(x, y[i])
```

```python
from matplotlib import cm
```

```python
cmap = cm.seismic
norm = colors.Normalize(vmin=A.min(), vmax=A.max())
```

```python
fig, ax= plt.subplots(1,1, figsize=(10,6))
for i in range(len(y)):
    ax.plot(x, y[i], color=cmap(norm(A[i])))

sm = cm.ScalarMappable(norm=norm, cmap=cmap)
cbar = fig.colorbar(sm, ax=ax)

ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$\sin(nx)$")
cbar.set_label(r"$n$")
```
