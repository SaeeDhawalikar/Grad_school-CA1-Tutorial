```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
samples = np.random.multivariate_normal([0,0], [[1,-0.75],[-0.75, 1]], 10000)

plt.scatter(samples[:,0], samples[:,1], s=1)
```

```python
from scipy.stats import gaussian_kde

# KDE fit
kde = gaussian_kde(samples.T)

# Grid for evaluation
grid_lim = 3
grid_pts = 100
x = np.linspace(-grid_lim, grid_lim, grid_pts)
y = np.linspace(-grid_lim, grid_lim, grid_pts)
X, Y = np.meshgrid(x, y)

# Evaluate KDE on the grid
positions = np.vstack([X.ravel(), Y.ravel()])
Z = kde(positions).reshape(X.shape)

# Plot contours
CS = plt.contourf(X, Y, Z, levels=10, cmap='Blues')
plt.colorbar(CS)
plt.show()
```

```python
np.savetxt("data/2D_gaussian.dat", Z, delimiter=',', fmt='%g')
```
