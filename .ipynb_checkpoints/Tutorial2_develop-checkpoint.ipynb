# Functions

### Univariate Gaussian

```python
# write a function to that generates a normalized 1-d Gaussian, given mu, sigma
# plot the output for a few combinations of mu,sigma. Plot a vertical line at the mean.
```

```python
def gauss(x, mu=0, sig=1):
    '''Gaussian function'''
    if (sig<=0):
        print("sigma must be positive definite")
        return -1
    y=(x-mu)**2/(2*sig**2)
    y=np.exp(-y)/np.sqrt(2*np.pi*sig**2)
    return y        
```

```python
x=np.arange(-3,3, 0.01)
y=gauss(x, 0, -1)
y
```

```python
mu1=1
sig1=2
x=np.arange(-10,10, 0.01)
y=gauss(x, mu1, sig1)
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, y,"k:")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.vlines(mu1, 0, 1/np.sqrt(2*np.pi*sig1**2), color="r", label=r"$\mu$")
plt.legend()
```

```python
np.random.seed(1)
N=int(1e4)
G=np.random.normal(loc=mu1, scale=sig1, size=N)
```

```python
plt.figure(figsize=(6,4))
plt.hist(G, bins=20, density=True, histtype="step", label="histogram")
plt.plot(x, y,"r", label="analytical")
plt.xlabel(r"$x$")
plt.ylabel("PDF")
plt.legend()
```

### sum of Gaussians

```python
def gauss_mix(x,mu, sig, A):
    y=np.exp(-(x-mu[:, None])**2/(2*sig[:,None]**2))
    y*=A[:,None]/np.sqrt(2*np.pi*sig[:,None]**2)
    y=np.sum(y, axis=0)
    return y
```

```python
def gauss_mix1(x,mu, sig, A):
    y=np.zeros_like(x)
    for i in range(len(mu)):
        y+=A[i]*gauss(x, mu[i], sig[i])
    return y 
```

```python
x=np.linspace(-5, 5,int(1e3))
mu=np.array([-3, 1])
sig=np.array([0.2, 1])
A=np.array([-1, 4])
```

```python
y1=gauss_mix(x, mu, sig, A)

y2=gauss_mix1(x, mu, sig, A)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, y1, "r")
plt.plot(x, y2, "k")
plt.xlabel("x")
plt.ylabel("y")

cols=["r", "b"]
plt.hlines(0, np.min(x), np.max(x), ls=":", color="grey")
for i in range(len(mu)):
    plt.vlines(mu[i], np.min(y1), np.max(y1), color=cols[i],label=r"$\mu_%d$"%i)
plt.legend()
```

```python
x=np.linspace(-5, 5,int(1e1))
ncomp=int(1e4)
mu=np.random.random(ncomp)
sig=np.random.random(ncomp)**2
A=np.random.random(ncomp)
```

```python
st=time.time()
y1=gauss_mix(x, mu, sig, A)
t1=time.time()-st
st=time.time()
y2=gauss_mix1(x, mu, sig, A)
t2=time.time()-st
```

### Multivariate Gaussian

```python
# Now, generate and plot a 2D gaussian, given the mu vector and the covariance matrix
# what are the constraints on the covariance matrix?
```

```python
x=np.random.random(1000)-0.5
y=(np.random.random(1000)-0.5)*2
X=np.array([x, y]).T
np.shape(X)
```

```python
mu=np.zeros(2)
sig=np.eye(2)
```

```python
def multi_gauss(X, mu=np.array([0,0]), cov=np.eye(2)):
    '''calculates the multivariate Gaussian PDF given:
    X: independent variables, shape[m,n], where n is the dimension, m number of points
    mu: mean array of the gaussian, shape[n]
    inv_cov: inverse covariance matrix, shape[n, n]
    returns:
    P: pdf at the given X values, shape [n, m]'''
    inv_cov=np.linalg.inv(cov)
    det_cov=np.linalg.det(cov)
    d=len(mu)
    A =1/np.sqrt((2*np.pi)**d*det_cov)
    y=np.matmul((X-mu), inv_cov)*(X-mu)
    y=-0.5*np.sum(y, axis=1)
    y=np.exp(y)*A
    return y
   
```

```python
P=multi_gauss(X, mu, sig)
np.shape(P)
```

```python
fig, ax= plt.subplots(1,1,figsize=(6,6))
im=ax.scatter(x, y, c=P)
ax.set_aspect(1)
ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$y$")
fig.colorbar(im, label="PDF")
```

```python
def multi_gauss_grid(X, mu=np.array([0,0]), cov=np.eye(2)):
    '''calculates the multivariate Gaussian PDF on a grid given:
    X: independent variables, shape[n,m], where n is the dimension, m number of points
    mu: mean array of the gaussian, shape[n]
    inv_cov: inverse covariance matrix, shape[n, n]
    returns:
    grids: grid on which the Gaussian is evaluated
    p: pdf at the given X values, shape [n, m]'''
    grids = np.meshgrid(*X, indexing="ij")
    points = np.stack([g.ravel() for g in grids], axis=-1)
    p = multi_gauss(points, mu, cov)
    p = p.reshape(grids[0].shape)
    return grids, p
```

```python
xmin=-2
xmax=2
ymin=-3
ymax=3
Nx=int(100)
Ny=int(100)
X=[np.linspace(xmin, xmax, Nx), np.linspace(ymin, ymax, Ny)]
cov=np.array([[1, -0.5], [-0.5 ,1]])
grids, pdf=multi_gauss_grid(X, mu, cov)
```

```python
np.min(X, axis=1), np.max(X, axis=1)
```

```python
np.shape(grids), np.shape(pdf)
```

```python
fig, ax= plt.subplots(1,2,figsize=(15,10))
im=ax[0].imshow(pdf.T, origin="lower", cmap="magma_r",
     extent=[xmin, xmax, ymin, ymax])
for i in range(2):
    ax[i].set_aspect(1)
    ax[i].set_xlabel(r"$x$")
    ax[i].set_ylabel(r"$y$")
fig.colorbar(im, ax=ax[0],label="PDF", shrink=0.85)
# notice: you have to do pdf.T

im=ax[1].contourf(grids[0], grids[1], pdf, cmap="magma_r")
fig.colorbar(im, ax=ax[1],label="PDF", shrink=0.85)
```

```python
xmin=-3
xmax=3
ymin=-3
ymax=3
Nx=int(100)
Ny=int(100)
X=[np.linspace(xmin, xmax, Nx), np.linspace(ymin, ymax, Ny)]
cov=np.array([[1, 0], [0 ,1]])
grids, pdf=multi_gauss_grid(X, mu, cov)
```
